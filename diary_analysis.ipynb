```python
# ==============================================================================
# 0. –ù–ê–°–¢–†–û–ô–ö–ê –°–†–ï–î–´ –ò –ó–ê–ì–†–£–ó–ö–ê –î–ê–ù–ù–´–•
# ==============================================================================
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# --- –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–π –¥–ª—è –ª—É—á—à–µ–≥–æ –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –≤ Colab ---
# –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Å—Ç–∏–ª—å –≥—Ä–∞—Ñ–∏–∫–æ–≤
sns.set_theme(style="whitegrid", rc={"figure.dpi": 100})
# –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –ø–∞–ª–∏—Ç—Ä—É
custom_palette = sns.color_palette("viridis", 10)
sns.set_palette(custom_palette)
# –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º —Ä–∞–∑–º–µ—Ä —à—Ä–∏—Ñ—Ç–∞
plt.rcParams.update({'font.size': 10})

# --- –ó–∞–≥—Ä—É–∑–∫–∞ –¥–∞–Ω–Ω—ã—Ö –∏–∑ —Ñ–∞–π–ª–∞ ---
file_path = 'revolution_events.json'
try:
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
    print(f"‚úÖ –§–∞–π–ª '{file_path}' —É—Å–ø–µ—à–Ω–æ –∑–∞–≥—Ä—É–∂–µ–Ω.")
except FileNotFoundError:
    print(f"‚ùå –û–®–ò–ë–ö–ê: –§–∞–π–ª '{file_path}' –Ω–µ –Ω–∞–π–¥–µ–Ω. –£–±–µ–¥–∏—Ç–µ—Å—å, —á—Ç–æ –≤—ã –∑–∞–≥—Ä—É–∑–∏–ª–∏ –µ–≥–æ –≤ Colab.")
    data = []

# --- –ü—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ –≤ Pandas DataFrame –¥–ª—è —É–¥–æ–±–Ω–æ–≥–æ –∞–Ω–∞–ª–∏–∑–∞ ---
if data:
    df = pd.json_normalize(data)
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –∫–æ–ª–æ–Ω–∫—É —Å –¥–∞—Ç–æ–π –≤ —Ñ–æ—Ä–º–∞—Ç datetime
    df['source_date'] = pd.to_datetime(df['source_date'])
    # –°–æ–∑–¥–∞–µ–º –∫–æ–ª–æ–Ω–∫—É —Å –º–µ—Å—è—Ü–µ–º –¥–ª—è –≥—Ä—É–ø–ø–∏—Ä–æ–≤–∫–∏
    df['month'] = df['source_date'].dt.to_period('M')
    # –°–æ–∑–¥–∞–µ–º –∫–æ–ª–æ–Ω–∫—É —Å –æ—Å–Ω–æ–≤–Ω–æ–π –∫–∞—Ç–µ–≥–æ—Ä–∏–µ–π –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞
    df['category'] = df['event_id'].apply(lambda x: x.split('_')[0] if isinstance(x, str) else 'UNKNOWN')

    print(f"‚úÖ –î–∞–Ω–Ω—ã–µ –ø—Ä–µ–æ–±—Ä–∞–∑–æ–≤–∞–Ω—ã –≤ DataFrame. –ó–∞–≥—Ä—É–∂–µ–Ω–æ {len(df)} —Å–æ–±—ã—Ç–∏–π.")
    print("-" * 50)
else:
    df = pd.DataFrame() # –°–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π DataFrame, —á—Ç–æ–±—ã –∏–∑–±–µ–∂–∞—Ç—å –æ—à–∏–±–æ–∫ –Ω–∏–∂–µ

# ==============================================================================
# 1. –û–ë–©–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê –ò –û–ë–ó–û–† –î–ê–ù–ù–´–•
# ==============================================================================
if not df.empty:
    print("üìä 1. –û–ë–©–ê–Ø –°–¢–ê–¢–ò–°–¢–ò–ö–ê –ò –û–ë–ó–û–† –î–ê–ù–ù–´–•")

    # –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π
    print(f"\n–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π: {len(df)}")

    # –í—Ä–µ–º–µ–Ω–Ω–æ–π –¥–∏–∞–ø–∞–∑–æ–Ω
    date_range_start = df['source_date'].min().strftime('%d %B %Y')
    date_range_end = df['source_date'].max().strftime('%d %B %Y')
    print(f"–í—Ä–µ–º–µ–Ω–Ω–æ–π –¥–∏–∞–ø–∞–∑–æ–Ω –¥–Ω–µ–≤–Ω–∏–∫–∞: —Å {date_range_start} –ø–æ {date_range_end}")

    # –†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ–±—ã—Ç–∏–π –ø–æ –æ—Å–Ω–æ–≤–Ω—ã–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
    category_names = {
        'REV1848': '–°–æ–±—ã—Ç–∏—è –≤ –ï–≤—Ä–æ–ø–µ',
        'RU': '–†–µ–∞–∫—Ü–∏—è –≤ –†–æ—Å—Å–∏–∏',
        'AUTHOR': '–í–æ—Å–ø—Ä–∏—è—Ç–∏–µ –∞–≤—Ç–æ—Ä–∞',
        'IDEOLOGIES': '–ò–¥–µ–æ–ª–æ–≥–∏–∏ –∏ –ø—Ä–∏—á–∏–Ω—ã',
        'OTHER': '–ü—Ä–æ—á–µ–µ'
    }
    df['category_name'] = df['category'].map(category_names).fillna('–ù–µ–∏–∑–≤–µ—Å—Ç–Ω–æ')

    category_counts = df['category_name'].value_counts()
    print("\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ–±—ã—Ç–∏–π –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º:")
    print(category_counts)

    # –ì—Ä–∞—Ñ–∏–∫ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –ø–æ –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º
    plt.figure(figsize=(10, 5))
    sns.barplot(x=category_counts.index, y=category_counts.values, palette="viridis")
    plt.title('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π –ø–æ –æ—Å–Ω–æ–≤–Ω—ã–º –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º')
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞–ø–∏—Å–µ–π')
    plt.xticks(rotation=15)
    plt.tight_layout()
    plt.show()

# ==============================================================================
# 2. –ê–ù–ê–õ–ò–ó –í–†–ï–ú–ï–ù–ù–´–• –†–Ø–î–û–í: –î–ò–ù–ê–ú–ò–ö–ê –°–û–ë–´–¢–ò–ô
# ==============================================================================
if not df.empty:
    print("\n" + "="*50)
    print("üóìÔ∏è 2. –ê–ù–ê–õ–ò–ó –í–†–ï–ú–ï–ù–ù–´–• –†–Ø–î–û–í: –î–ò–ù–ê–ú–ò–ö–ê –°–û–ë–´–¢–ò–ô")

    # --- –ì—Ä—É–ø–ø–∏—Ä—É–µ–º –¥–∞–Ω–Ω—ã–µ –ø–æ –Ω–µ–¥–µ–ª—è–º (–æ–¥–∏–Ω —Ä–∞–∑ –¥–ª—è –æ–±–æ–∏—Ö –≥—Ä–∞—Ñ–∏–∫–æ–≤) ---
    # W-Mon –æ–∑–Ω–∞—á–∞–µ—Ç, —á—Ç–æ –Ω–µ–¥–µ–ª—è –Ω–∞—á–∏–Ω–∞–µ—Ç—Å—è —Å –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫–∞
    events_per_week = df.resample('W-Mon', on='source_date').size()

    # --- START: –û–ë–ù–û–í–õ–ï–ù–ù–´–ô –ë–õ–û–ö - –°–¢–û–õ–ë–ß–ê–¢–ê–Ø –î–ò–ê–ì–†–ê–ú–ú–ê –ü–û –ù–ï–î–ï–õ–Ø–ú ---

    # –§–æ—Ä–º–∞—Ç–∏—Ä—É–µ–º –Ω–∞–∑–≤–∞–Ω–∏—è –Ω–µ–¥–µ–ª—å –¥–ª—è –æ—Å–∏ X
    weekly_labels = []
    for week_start_date in events_per_week.index:
        week_end_date = week_start_date + pd.Timedelta(days=6)
        # –§–æ—Ä–º–∞—Ç–∏—Ä—É–µ–º –¥–∞—Ç—É –≤ DD.MM
        start_str = week_start_date.strftime('%d.%m')
        end_str = week_end_date.strftime('%d.%m')
        weekly_labels.append(f"{start_str}‚Äì{end_str}")

    print("\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ–±—ã—Ç–∏–π –ø–æ –Ω–µ–¥–µ–ª—è–º:")
    # –í—ã–≤–æ–¥–∏–º –¥–∞–Ω–Ω—ã–µ –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏
    for label, count in zip(weekly_labels, events_per_week.values):
        print(f"–ù–µ–¥–µ–ª—è {label}: {count} —Å–æ–±—ã—Ç–∏–π")

    # –°—Ç—Ä–æ–∏–º —Å—Ç–æ–ª–±—á–∞—Ç—É—é –¥–∏–∞–≥—Ä–∞–º–º—É
    plt.figure(figsize=(14, 7))
    bars = sns.barplot(x=weekly_labels, y=events_per_week.values, palette="viridis")
    plt.title('–û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π –ø–æ –Ω–µ–¥–µ–ª—è–º –≤ 1849 –≥–æ–¥—É')
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π')
    plt.xlabel('–ù–µ–¥–µ–ª—è (–ù–∞—á–∞–ª–æ‚Äì–ö–æ–Ω–µ—Ü)')
    plt.xticks(rotation=90, fontsize=8) # –ü–æ–≤–æ—Ä–∞—á–∏–≤–∞–µ–º –º–µ—Ç–∫–∏ –Ω–∞ 90 –≥—Ä–∞–¥—É—Å–æ–≤ –¥–ª—è —á–∏—Ç–∞–µ–º–æ—Å—Ç–∏

    # –î–æ–±–∞–≤–ª—è–µ–º –ø–æ–¥–ø–∏—Å–∏ —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –Ω–∞–¥ –∫–∞–∂–¥—ã–º —Å—Ç–æ–ª–±—Ü–æ–º
    for bar in bars.patches:
        bars.annotate(format(bar.get_height(), '.0f'),
                      (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                      ha='center', va='bottom',
                      size=8, xytext=(0, 2),
                      textcoords='offset points')

    plt.tight_layout()
    plt.show()

    # --- END: –û–ë–ù–û–í–õ–ï–ù–ù–´–ô –ë–õ–û–ö ---

    # –õ–∏–Ω–µ–π–Ω—ã–π –≥—Ä–∞—Ñ–∏–∫ (–∏—Å–ø–æ–ª—å–∑—É–µ–º —Ç–µ –∂–µ –¥–∞–Ω–Ω—ã–µ, —á—Ç–æ –∏ –¥–ª—è —Å—Ç–æ–ª–±—á–∞—Ç–æ–π –¥–∏–∞–≥—Ä–∞–º–º—ã)
    plt.figure(figsize=(14, 6))
    events_per_week.plot(kind='line', marker='o', linestyle='-')
    plt.title('–î–∏–Ω–∞–º–∏–∫–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–∞ —Å–æ–±—ã—Ç–∏–π –ø–æ –Ω–µ–¥–µ–ª—è–º (1849 –≥–æ–¥)')
    plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ–±—ã—Ç–∏–π')
    plt.xlabel('–î–∞—Ç–∞ –Ω–∞—á–∞–ª–∞ –Ω–µ–¥–µ–ª–∏')
    plt.grid(True, which='both', linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.show()

    # –ê–Ω–∞–ª–∏–∑ —Ñ–æ–∫—É—Å–∞ –≤–Ω–∏–º–∞–Ω–∏—è –∞–≤—Ç–æ—Ä–∞ –ø–æ –º–µ—Å—è—Ü–∞–º (–æ—Å—Ç–∞–µ—Ç—Å—è –±–µ–∑ –∏–∑–º–µ–Ω–µ–Ω–∏–π)
    monthly_focus = df.groupby('month')['category_name'].value_counts(normalize=True).unstack().fillna(0)

    ax = monthly_focus.plot(kind='bar', stacked=True, figsize=(14, 7), colormap='viridis')
    plt.title('–ò–∑–º–µ–Ω–µ–Ω–∏–µ —Ñ–æ–∫—É—Å–∞ –≤–Ω–∏–º–∞–Ω–∏—è –∞–≤—Ç–æ—Ä–∞ –ø–æ –º–µ—Å—è—Ü–∞–º')
    plt.ylabel('–î–æ–ª—è —É–ø–æ–º–∏–Ω–∞–Ω–∏–π (%)')
    plt.xlabel('–ú–µ—Å—è—Ü')
    plt.xticks(rotation=45)
    plt.legend(title='–ö–∞—Ç–µ–≥–æ—Ä–∏—è', bbox_to_anchor=(1.02, 1), loc='upper left')
    ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
    plt.tight_layout()
    plt.show()

# ==============================================================================
# 3. –ì–ï–û–ì–†–ê–§–ò–ß–ï–°–ö–ò–ô –ê–ù–ê–õ–ò–ó
# ==============================================================================
if not df.empty:
    print("\n" + "="*50)
    print("üó∫Ô∏è 3. –ì–ï–û–ì–†–ê–§–ò–ß–ï–°–ö–ò–ô –ê–ù–ê–õ–ò–ó")

    location_counts = df['location_normalized'].value_counts().nlargest(10)
    print("\n–¢–æ–ø-10 –ª–æ–∫–∞—Ü–∏–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —É–ø–æ–º–∏–Ω–∞–Ω–∏–π:")
    print(location_counts)

    plt.figure(figsize=(10, 6))
    sns.barplot(y=location_counts.index, x=location_counts.values, orient='h')
    plt.title('–¢–æ–ø-10 –ª–æ–∫–∞—Ü–∏–π –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —É–ø–æ–º–∏–Ω–∞–Ω–∏–π')
    plt.xlabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–ø–æ–º–∏–Ω–∞–Ω–∏–π')
    plt.ylabel('–õ–æ–∫–∞—Ü–∏—è')
    plt.tight_layout()
    plt.show()

    # –ë–æ–ª–µ–µ –≥–ª—É–±–æ–∫–∏–π –∞–Ω–∞–ª–∏–∑: –∫–∞–∫–∏–µ —Ç–µ–º—ã –æ–±—Å—É–∂–¥–∞—é—Ç—Å—è –≤ –∫–ª—é—á–µ–≤—ã—Ö –ª–æ–∫–∞—Ü–∏—è—Ö?
    top_locations = location_counts.index
    location_category_crosstab = pd.crosstab(
        df[df['location_normalized'].isin(top_locations)]['location_normalized'],
        df['category_name']
    )
    print("\n–¢–µ–º–∞—Ç–∏–∫–∞ —Å–æ–±—ã—Ç–∏–π –≤ –∫–ª—é—á–µ–≤—ã—Ö –ª–æ–∫–∞—Ü–∏—è—Ö:")
    print(location_category_crosstab)

# ==============================================================================
# 4. –ê–ù–ê–õ–ò–ó –ò–°–¢–û–ß–ù–ò–ö–û–í –ò–ù–§–û–†–ú–ê–¶–ò–ò
# ==============================================================================
if not df.empty:
    print("\n" + "="*50)
    print("üì∞ 4. –ê–ù–ê–õ–ò–ó –ò–°–¢–û–ß–ù–ò–ö–û–í –ò–ù–§–û–†–ú–ê–¶–ò–ò")

    source_type_counts = df['information_source_type'].value_counts()

    plt.figure(figsize=(8, 8))
    plt.pie(source_type_counts, labels=source_type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel", len(source_type_counts)))
    plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Å–æ–±—ã—Ç–∏–π –ø–æ —Ç–∏–ø—É –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏')
    plt.ylabel('')
    plt.show()

    # –ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –º–µ–∂–¥—É –∏—Å—Ç–æ—á–Ω–∏–∫–æ–º –∏ —Ç–∏–ø–æ–º —Ä–µ–∞–∫—Ü–∏–∏ –∞–≤—Ç–æ—Ä–∞
    author_df = df[df['category'] == 'AUTHOR'].copy()
    author_df['perception_type'] = author_df['event_name'].apply(lambda x: x.split('(')[0].strip())

    source_perception_crosstab = pd.crosstab(
        author_df['information_source_type'],
        author_df['perception_type']
    )
    print("\n–°–≤—è–∑—å –º–µ–∂–¥—É –∏—Å—Ç–æ—á–Ω–∏–∫–æ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ –∏ —Ç–∏–ø–æ–º —Ä–µ–∞–∫—Ü–∏–∏ –∞–≤—Ç–æ—Ä–∞:")
    print(source_perception_crosstab)

    plt.figure(figsize=(12, 7))
    sns.heatmap(source_perception_crosstab, annot=True, fmt='d', cmap='viridis')
    plt.title('–¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞: –ò—Å—Ç–æ—á–Ω–∏–∫ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–∏ vs. –¢–∏–ø —Ä–µ–∞–∫—Ü–∏–∏ –∞–≤—Ç–æ—Ä–∞')
    plt.ylabel('–¢–∏–ø –∏—Å—Ç–æ—á–Ω–∏–∫–∞')
    plt.xlabel('–¢–∏–ø —Ä–µ–∞–∫—Ü–∏–∏')
    plt.xticks(rotation=20, ha='right')
    plt.tight_layout()
    plt.show()

# ==============================================================================
# 5. –•–†–û–ù–û–õ–û–ì–ò–Ø –≠–ú–û–¶–ò–ô –ò –û–¢–ù–û–®–ï–ù–ò–Ø –ê–í–¢–û–†–ê
# ==============================================================================
if not df.empty:
    print("\n" + "="*50)
    print("üé≠ 5. –•–†–û–ù–û–õ–û–ì–ò–Ø –≠–ú–û–¶–ò–ô –ò –û–¢–ù–û–®–ï–ù–ò–Ø –ê–í–¢–û–†–ê")

    # –§–∏–ª—å—Ç—Ä—É–µ–º —Å–æ–±—ã—Ç–∏—è, –æ—Ç—Ä–∞–∂–∞—é—â–∏–µ —ç–º–æ—Ü–∏–∏ –∏–ª–∏ —Å–∏–ª—å–Ω–æ–µ –ª–∏—á–Ω–æ–µ –º–Ω–µ–Ω–∏–µ
    emotion_keywords = ['—Å—Ç—Ä–∞—Ö', '—Ç—Ä–µ–≤–æ–≥–∞', '–Ω–µ–Ω–∞–≤–∏—Å—Ç—å', '–≥–Ω–µ–≤', '—Ä–∞–¥–æ—Å—Ç—å', '–≤–æ—Å—Ç–æ—Ä–≥', '–ª—é–±–æ–ø—ã—Ç—Å—Ç–≤–æ', '–∏–Ω—Ç–µ—Ä–µ—Å']
    emotion_df = df[df['event_id'].str.contains('_EMO_') | df['event_name'].str.contains('–†–∞–∑–º—ã—à–ª–µ–Ω–∏—è')].copy()

    print(f"\n–ù–∞–π–¥–µ–Ω–æ {len(emotion_df)} –∑–∞–ø–∏—Å–µ–π, –æ—Ç—Ä–∞–∂–∞—é—â–∏—Ö —ç–º–æ—Ü–∏–∏ –∏ —Ä–∞–∑–º—ã—à–ª–µ–Ω–∏—è –∞–≤—Ç–æ—Ä–∞.")
    print("\n–ü—Ä–∏–º–µ—Ä—ã –∑–∞–ø–∏—Å–µ–π:")

    # –í—ã–≤–æ–¥–∏–º –Ω–µ—Å–∫–æ–ª—å–∫–æ –ø—Ä–∏–º–µ—Ä–æ–≤ –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü—ã –¥–ª—è –Ω–∞–≥–ª—è–¥–Ω–æ—Å—Ç–∏
    emotion_df['date_str'] = emotion_df['source_date'].dt.strftime('%Y-%m-%d')
    emotion_preview = emotion_df[['date_str', 'event_subtype_custom', 'text_fragment']].head(10)

    # –î–ª—è –∫—Ä–∞—Å–∏–≤–æ–≥–æ –≤—ã–≤–æ–¥–∞ –≤ Colab
    from IPython.display import display
    display(emotion_preview)

# ==============================================================================
# 6. –ê–ù–ê–õ–ò–ó –ö–õ–Æ–ß–ï–í–´–• –°–õ–û–í (–û–ë–õ–ê–ö–û –¢–ï–ì–û–í)
# ==============================================================================
if not df.empty:
    print("\n" + "="*50)
    print("‚òÅÔ∏è 6. –ê–ù–ê–õ–ò–ó –ö–õ–Æ–ß–ï–í–´–• –°–õ–û–í")

    # –£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫—É –¥–ª—è –æ–±–ª–∞–∫–∞ —Å–ª–æ–≤, –µ—Å–ª–∏ –µ–µ –Ω–µ—Ç
    try:
        from wordcloud import WordCloud
    except ImportError:
        print("–£—Å—Ç–∞–Ω–∞–≤–ª–∏–≤–∞–µ–º –±–∏–±–ª–∏–æ—Ç–µ–∫—É wordcloud...")
        !pip install wordcloud
        from wordcloud import WordCloud

    # –°–æ–±–∏—Ä–∞–µ–º –≤—Å–µ –∫–ª—é—á–µ–≤—ã–µ —Å–ª–æ–≤–∞ –≤ –æ–¥–∏–Ω —Å–ø–∏—Å–æ–∫
    all_keywords = [keyword for sublist in df['keywords'].dropna() for keyword in sublist]

    if all_keywords:
        word_counts = Counter(all_keywords)
        print(f"\n–¢–æ–ø-20 —Å–∞–º—ã—Ö —á–∞—Å—Ç—ã—Ö –∫–ª—é—á–µ–≤—ã—Ö —Å–ª–æ–≤:")
        for word, count in word_counts.most_common(20):
            print(f"- {word}: {count}")

        # –°–æ–∑–¥–∞–Ω–∏–µ –æ–±–ª–∞–∫–∞ —Å–ª–æ–≤
        wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate_from_frequencies(word_counts)

        plt.figure(figsize=(15, 7))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('–û–±–ª–∞–∫–æ –∫–ª—é—á–µ–≤—ã—Ö —Å–ª–æ–≤ –ø–æ –≤—Å–µ–º—É –¥–Ω–µ–≤–Ω–∏–∫—É')
        plt.show()
    else:
        print("\n–ö–ª—é—á–µ–≤—ã–µ —Å–ª–æ–≤–∞ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω—ã.")

# ==============================================================================
# 7. –ò–¢–û–ì–û–í–´–ï –í–´–í–û–î–´ –î–õ–Ø –ò–°–°–õ–ï–î–û–í–ê–ù–ò–Ø
# ==============================================================================
if not df.empty:
    print("\n" + "="*50)
    print("üî¨ 7. –ö–õ–Æ–ß–ï–í–´–ï –í–´–í–û–î–´ –î–õ–Ø –ò–°–¢–û–†–ò–ß–ï–°–ö–û–ì–û –ò–°–°–õ–ï–î–û–í–ê–ù–ò–Ø")

    print("""
    1.  **–î–≤–æ–π—Å—Ç–≤–µ–Ω–Ω–æ—Å—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏–æ–Ω–Ω–æ–≥–æ –ø–æ–ª—è:** –ê–Ω–∞–ª–∏–∑ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —á–µ—Ç–∫–æ–µ —Ä–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –¥–≤–∞ –º–∏—Ä–∞: –ª–æ–∫–∞–ª—å–Ω—ã–π (–í–æ–ª–æ–≥–¥–∞) –∏ –≥–ª–æ–±–∞–ª—å–Ω—ã–π (–ï–≤—Ä–æ–ø–∞). –í –í–æ–ª–æ–≥–¥–µ –∞–≤—Ç–æ—Ä–∞ –±–µ—Å–ø–æ–∫–æ—è—Ç —É—Å–∏–ª–µ–Ω–∏–µ –ø–æ–ª–∏—Ü–µ–π—Å–∫–æ–≥–æ –Ω–∞–¥–∑–æ—Ä–∞ –∏ —Ä–µ—Ñ–æ—Ä–º—ã –æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏—è. –ù–æ–≤–æ—Å—Ç–∏ –∏–∑ –ï–≤—Ä–æ–ø—ã, –ø–æ–ª—É—á–∞–µ–º—ã–µ –∏–∑ –≥–∞–∑–µ—Ç, —Ñ–æ—Ä–º–∏—Ä—É—é—Ç –µ–≥–æ –∫–∞—Ä—Ç–∏–Ω—É "–í–µ—Å–Ω—ã –Ω–∞—Ä–æ–¥–æ–≤".

    2.  **–î–∏–Ω–∞–º–∏–∫–∞ –∏–Ω—Ç–µ—Ä–µ—Å–∞:** –í—Ä–µ–º–µ–Ω–Ω–æ–π –∞–Ω–∞–ª–∏–∑ –¥–µ–º–æ–Ω—Å—Ç—Ä–∏—Ä—É–µ—Ç, —á—Ç–æ –∏–Ω—Ç–µ—Ä–µ—Å –∞–≤—Ç–æ—Ä–∞ –∫ –µ–≤—Ä–æ–ø–µ–π—Å–∫–∏–º —Å–æ–±—ã—Ç–∏—è–º –Ω–µ –±—ã–ª –ø–æ—Å—Ç–æ—è–Ω–Ω—ã–º. –í—Å–ø–ª–µ—Å–∫–∏ –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç–∏ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ —á–µ—Ç–∫–æ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—Ç —Å –ø–æ–ª—É—á–µ–Ω–∏–µ–º –Ω–æ–≤–æ—Å—Ç–µ–π –æ –∫–ª—é—á–µ–≤—ã—Ö —ç—Ç–∞–ø–∞—Ö –í–µ–Ω–≥–µ—Ä—Å–∫–æ–π –∫–∞–º–ø–∞–Ω–∏–∏, —á—Ç–æ –≥–æ–≤–æ—Ä–∏—Ç –æ —Å–æ–±—ã—Ç–∏–π–Ω–æ-–æ—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–∞–Ω–Ω–æ–º, –∞ –Ω–µ –ø–æ—Å—Ç–æ—è–Ω–Ω–æ–º —Ñ–æ–Ω–æ–≤–æ–º –∏–Ω—Ç–µ—Ä–µ—Å–µ.

    3.  **–í–ª–∏—è–Ω–∏–µ –∏—Å—Ç–æ—á–Ω–∏–∫–æ–≤:** –¢–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç —Å–∏–ª—å–Ω—É—é —Å–≤—è–∑—å –º–µ–∂–¥—É —Ç–∏–ø–æ–º –∏—Å—Ç–æ—á–Ω–∏–∫–∞ –∏ —Ä–µ–∞–∫—Ü–∏–µ–π. "–û—Ñ–∏—Ü–∏–∞–ª—å–Ω—ã–µ –∏—Å—Ç–æ—á–Ω–∏–∫–∏ (–≥–∞–∑–µ—Ç—ã)" —á–∞—Å—Ç–æ –ø–æ—Ä–æ–∂–¥–∞—é—Ç –∑–∞–ø–∏—Å–∏ —Å "–ü–æ–¥–¥–µ—Ä–∂–∫–æ–π –¥–µ–π—Å—Ç–≤–∏–π –≤–ª–∞—Å—Ç–µ–π", –≤ —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ "–°–ª—É—Ö–∏" –∏ "–õ–∏—á–Ω—ã–µ –Ω–∞–±–ª—é–¥–µ–Ω–∏—è" —Å–≤—è–∑–∞–Ω—ã —Å –±–æ–ª–µ–µ —à–∏—Ä–æ–∫–∏–º —Å–ø–µ–∫—Ç—Ä–æ–º —Ä–µ–∞–∫—Ü–∏–π, –≤–∫–ª—é—á–∞—è —Ç—Ä–µ–≤–æ–≥—É –∏ —Ä–∞–∑–º—ã—à–ª–µ–Ω–∏—è –æ –±—É–¥—É—â–µ–º.

    4.  **–≠–º–æ—Ü–∏–æ–Ω–∞–ª—å–Ω—ã–π –ª–∞–Ω–¥—à–∞—Ñ—Ç:** –•—Ä–æ–Ω–æ–ª–æ–≥–∏—è —ç–º–æ—Ü–∏–π –≤—ã—è–≤–ª—è–µ—Ç –ø—Ä–µ–æ–±–ª–∞–¥–∞–Ω–∏–µ –∫–æ–Ω—Å–µ—Ä–≤–∞—Ç–∏–≤–Ω—ã—Ö –∏ –ø–∞—Ç—Ä–∏–æ—Ç–∏—á–µ—Å–∫–∏—Ö –Ω–∞—Å—Ç—Ä–æ–µ–Ω–∏–π (–ø–æ–¥–¥–µ—Ä–∂–∫–∞ –º–æ–Ω–∞—Ä—Ö–∏–∏, –∞–Ω—Ç–∏–ø–æ–ª—å—Å–∫–∏–µ –≤—ã–ø–∞–¥—ã), –Ω–æ —Ç–∞–∫–∂–µ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –º–æ–º–µ–Ω—Ç—ã –ª–∏—á–Ω–æ–π —Ç—Ä–µ–≤–æ–≥–∏, —Å—Ç—Ä–∞—Ö–∞ –ø–µ—Ä–µ–¥ –±—É–¥—É—â–∏–º –∏ —Ä–µ—Ñ–ª–µ–∫—Å–∏–∏ –æ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—É–¥—å–±–µ –≤ –∫–æ–Ω—Ç–µ–∫—Å—Ç–µ –≥–ª–æ–±–∞–ª—å–Ω—ã—Ö –ø–æ—Ç—Ä—è—Å–µ–Ω–∏–π.

    5.  **–ö–ª—é—á–µ–≤—ã–µ —Ç–µ–º—ã:** –û–±–ª–∞–∫–æ —Å–ª–æ–≤ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç, —á—Ç–æ —Ü–µ–Ω—Ç—Ä–∞–ª—å–Ω—ã–º–∏ —Ç–µ–º–∞–º–∏ –¥–ª—è –∞–≤—Ç–æ—Ä–∞ –≤ 1849 –≥–æ–¥—É –±—ã–ª–∏ "–≤–æ–π–Ω–∞", "–í–µ–Ω–≥—Ä–∏—è", "–ø–æ–ª—è–∫–∏", "—Ü–µ–Ω–∑—É—Ä–∞" –∏ "–æ–±—Ä–∞–∑–æ–≤–∞–Ω–∏–µ". –≠—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –æ–±—ä–µ–∫—Ç–∏–≤–Ω–æ –æ—Ü–µ–Ω–∏—Ç—å –µ–≥–æ –ø—Ä–∏–æ—Ä–∏—Ç–µ—Ç—ã –∏ –∫—Ä—É–≥ –∏–Ω—Ç–µ—Ä–µ—Å–æ–≤.
    """)
```
