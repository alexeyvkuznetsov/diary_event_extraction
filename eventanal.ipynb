```python
# -*- coding: utf-8 -*-
"""
Анализ данных из revolution_events.json (версия без author_sentiment и perception_target).

Этот скрипт загружает структурированные данные, извлеченные из дневника
К.А. Березкина, и проводит предварительный количественный анализ
для выявления основных тем, источников информации и т.д.
"""

import json
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates # Для форматирования дат на оси X
import seaborn as sns
from collections import Counter
from wordcloud import WordCloud
import os

# --- Настройки ---
# Укажите путь к вашему файлу JSON на Google Drive
# Пример: '/content/drive/MyDrive/Colab Notebooks/revolution_events.json'
# ВАЖНО: Измените этот путь на ваш реальный путь!
JSON_FILE_PATH = '/content/revolution_events.json'

# Путь для сохранения визуализаций (папка будет создана, если не существует)
PLOTS_SAVE_PATH = '/content/revolution_plots_simplified/'

# Количество топ-элементов для отображения на графиках
TOP_N = 15

# Настройки визуализации
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 7) # Размер графиков по умолчанию
plt.rcParams['font.family'] = 'DejaVu Sans' # Шрифт, поддерживающий кириллицу (обычно доступен в Colab)

# --- Шаг 1: Подключение Google Drive и Загрузка данных ---

try:
    from google.colab import drive
    drive.mount('/content/drive')
    print("Google Drive подключен.")
except ImportError:
    print("Скрипт запущен не в Google Colab или модуль google.colab не найден.")
    # Если запускаете локально, закомментируйте строки выше и укажите локальный путь
    # JSON_FILE_PATH = 'path/to/your/local/revolution_events.json'
    # PLOTS_SAVE_PATH = './revolution_plots_simplified/' # Сохранять в текущую папку

# Создаем папку для графиков, если ее нет
if not os.path.exists(PLOTS_SAVE_PATH):
    os.makedirs(PLOTS_SAVE_PATH)
    print(f"Папка для графиков создана: {PLOTS_SAVE_PATH}")

try:
    with open(JSON_FILE_PATH, 'r', encoding='utf-8') as f:
        data = json.load(f)
    df = pd.DataFrame(data)
    print(f"Данные успешно загружены. Количество записей: {len(df)}")
    print("Первые 5 строк данных:")
    print(df.head())
except FileNotFoundError:
    print(f"Ошибка: Файл не найден по пути: {JSON_FILE_PATH}")
    exit()
except json.JSONDecodeError:
    print(f"Ошибка: Не удалось декодировать JSON из файла: {JSON_FILE_PATH}")
    exit()
except Exception as e:
    print(f"Произошла ошибка при загрузке данных: {e}")
    exit()

# --- Шаг 2: Предобработка данных ---

# Преобразование дат
df['source_date'] = pd.to_datetime(df['source_date'], errors='coerce')
# Удаление строк с некорректными датами, если они есть
df.dropna(subset=['source_date'], inplace=True)

print("\nТипы данных после преобразования:")
print(df.info())

# --- Шаг 3: Анализ и Визуализация (адаптировано под новую структуру) ---

# 3.1. Общая Характеристика Набора Данных (Распределение по event_id)
print(f"\n--- Анализ: Распределение по Event ID (Топ {TOP_N}) ---")
event_id_counts = df['event_id'].value_counts().head(TOP_N)
print(event_id_counts)

plt.figure(figsize=(12, 8))
sns.barplot(y=event_id_counts.index, x=event_id_counts.values, palette='viridis')
plt.title(f'Топ {TOP_N} наиболее часто встречающихся Event ID')
plt.xlabel('Количество упоминаний')
plt.ylabel('Event ID')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_SAVE_PATH, 'top_event_ids.png'))
plt.show()

# 3.2. Тематический Фокус (Распределение по event_name)
print(f"\n--- Анализ: Распределение по Event Name (Топ {TOP_N}) ---")
event_name_counts = df['event_name'].value_counts().head(TOP_N)
print(event_name_counts)

plt.figure(figsize=(12, 8))
sns.barplot(y=event_name_counts.index, x=event_name_counts.values, palette='viridis')
plt.title(f'Топ {TOP_N} наиболее часто встречающихся Названий Событий')
plt.xlabel('Количество упоминаний')
plt.ylabel('Название События')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_SAVE_PATH, 'top_event_names.png'))
plt.show()

# 3.2.1. Анализ по event_subtype_custom (если есть значимые данные)
if 'event_subtype_custom' in df.columns and df['event_subtype_custom'].notna().any():
    print(f"\n--- Анализ: Распределение по Event Subtype Custom (Топ {TOP_N}) ---")
    subtype_counts = df['event_subtype_custom'].dropna().value_counts().head(TOP_N)
    if not subtype_counts.empty:
        print(subtype_counts)
        plt.figure(figsize=(12, 8))
        sns.barplot(y=subtype_counts.index, x=subtype_counts.values, palette='mako')
        plt.title(f'Топ {TOP_N} Event Subtype Custom')
        plt.xlabel('Количество упоминаний')
        plt.ylabel('Уточнение типа события')
        plt.tight_layout()
        plt.savefig(os.path.join(PLOTS_SAVE_PATH, 'top_event_subtypes.png'))
        plt.show()
    else:
        print("Значимых данных для 'event_subtype_custom' не найдено.")
else:
    print("Поле 'event_subtype_custom' отсутствует или пусто.")


# --- ИЗМЕНЕННЫЙ БЛОК: 3.3. Динамика Упоминаний во времени (по неделям) ---
print("\n--- Анализ: Динамика упоминаний по неделям ---")
df_time = df.set_index('source_date')
# Группируем по неделям (начало недели - понедельник) и считаем количество записей
# 'W' или 'W-MON' для группировки по неделям, начинающимся с понедельника
weekly_counts = df_time.resample('W-MON').size()
print(weekly_counts)

# Линейный график
plt.figure(figsize=(18, 8)) # Увеличим размер для лучшей читаемости недельных данных
weekly_counts.plot(kind='line', marker='o', linestyle='-')
plt.title('Динамика упоминаний событий/восприятия по неделям (1849 г.)')
plt.xlabel('Неделя (Начало недели)')
plt.ylabel('Количество упоминаний')
plt.grid(True)
# Форматирование оси X для отображения дат в более читаемом виде
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=1)) # Метки каждый месяц
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_SAVE_PATH, 'weekly_dynamics_line.png'))
plt.show()

# Столбчатая диаграмма
plt.figure(figsize=(18, 8))
weekly_counts.plot(kind='bar', color=sns.color_palette("viridis", len(weekly_counts)))
plt.title('Динамика упоминаний событий/восприятия по неделям (1849 г.) - Столбчатая диаграмма')
plt.xlabel('Неделя (Начало недели)')
plt.ylabel('Количество упоминаний')
plt.grid(axis='y') # Сетка только по оси Y для столбчатой диаграммы
# Форматирование оси X для отображения дат в более читаемом виде
# Для bar plot индексы - это уже строки, если resample вернул их так, либо Timestamp
# Если weekly_counts.index это Timestamp, то форматируем:
if pd.api.types.is_datetime64_any_dtype(weekly_counts.index):
    labels = weekly_counts.index.strftime('%Y-%m-%d')
else:
    labels = weekly_counts.index
plt.xticks(ticks=range(len(labels)), labels=labels, rotation=45, ha='right', fontsize=8) # Уменьшим шрифт для меток
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_SAVE_PATH, 'weekly_dynamics_bar.png'))
plt.show()
# --- КОНЕЦ ИЗМЕНЕННОГО БЛОКА ---


# 3.4. Источники Информации
print("\n--- Анализ: Распределение по Типу Источника Информации ---")
source_type_counts = df['information_source_type'].fillna('Источник неясен/не указан').value_counts()
print(source_type_counts)

plt.figure(figsize=(12, 7))
sns.barplot(x=source_type_counts.index, y=source_type_counts.values, palette='coolwarm')
plt.title('Распределение по Типу Источника Информации')
plt.xlabel('Тип Источника')
plt.ylabel('Количество упоминаний')
plt.xticks(rotation=25, ha='right')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_SAVE_PATH, 'source_type_bar.png'))
plt.show()

# 3.5. Географический Аспект
print(f"\n--- Анализ: Распределение по Нормализованным Локациям (Топ {TOP_N}) ---")
location_counts = df['location_normalized'].fillna('Не указано').value_counts().head(TOP_N)
print(location_counts)

plt.figure(figsize=(12, 8))
sns.barplot(y=location_counts.index, x=location_counts.values, palette='crest')
plt.title(f'Топ {TOP_N} Нормализованных Локаций Событий')
plt.xlabel('Количество упоминаний')
plt.ylabel('Локация')
plt.tight_layout()
plt.savefig(os.path.join(PLOTS_SAVE_PATH, 'top_locations.png'))
plt.show()

# 3.6. Анализ Ключевых Слов (Word Cloud)
print("\n--- Анализ: Облако Ключевых Слов ---")
try:
    all_keywords = [kw.lower() for sublist in df['keywords'].dropna() for kw in sublist if isinstance(kw, str)]

    if all_keywords:
        keyword_counts = Counter(all_keywords)
        print(f"Топ 30 Ключевых Слов:")
        print(keyword_counts.most_common(30))

        wordcloud = WordCloud(width=1000, height=500, background_color='white',
                              colormap='tab10', max_words=100).generate_from_frequencies(keyword_counts)

        plt.figure(figsize=(15, 8))
        plt.imshow(wordcloud, interpolation='bilinear')
        plt.axis('off')
        plt.title('Облако Ключевых Слов')
        plt.tight_layout(pad=0)
        plt.savefig(os.path.join(PLOTS_SAVE_PATH, 'keywords_wordcloud.png'))
        plt.show()
    else:
        print("Ключевые слова не найдены или некорректны для создания облака.")

except ImportError:
    print("Библиотека 'wordcloud' не установлена. Пропустите этот шаг или установите ее: pip install wordcloud")
except Exception as e:
    print(f"Произошла ошибка при создании облака слов: {e}")

# --- Завершение ---
print("\nАнализ завершен. Графики сохранены в папку:", PLOTS_SAVE_PATH)
```
